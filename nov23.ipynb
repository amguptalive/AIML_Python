```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

def show_image(img, title="Image"):
    plt.imshow(img)
    plt.title(title)
    plt.axis('off')
    plt.show()
    print(f"Width: {len(img[0])}, Height: {len(img)}, RGB: {img[0][0]}")

def enhance_red(image_path, red_factor=1.5, green_factor=0.7, blue_factor=0.7):
    """
    Enhances the red component of an image while reducing green and blue components.

    Args:
        image_path: Path to the image file.
        red_factor: Factor to multiply the red component.
        green_factor: Factor to multiply the green component.
        blue_factor: Factor to multiply the blue component.

    Returns:
        The modified image.
    """
    img = mpimg.imread(image_path)

    # Make a writable copy of the image
    img = img.copy()

    height, width, channels = img.shape

    for y in range(height):
        for x in range(width):
            r, g, b = img[y, x]
            r = min(int(r * red_factor), 255)
            g = max(int(g * green_factor), 0)
            b = max(int(b * blue_factor), 0)
            img[y, x] = [r, g, b]

    return img

# Example usage:
image_path = "istockphoto-2155904673-1024x1024.jpg"  
myimg = mpimg.imread(image_path)

# Display the original and enhanced images
show_image(myimg, "Original Image")
enhanced_image = enhance_red(image_path, red_factor=2.0, green_factor=0.5, blue_factor=0.5)
show_image(enhanced_image, "Enhanced Image")
```

```python
'''This is a constructor'''
class Vector:
    # it does something
    '''This is a constructor'''
    def __init__(self, x, y):
        if type(x) != int:
            print("ERROR!")
            return None
        self.x = x
        self.y = y
    def norm(self):
        x, y = self.x, self.y
        return math.sqrt(x*x + y * y)
    def add(self, other):
        xx = self.x + other.x
        yy = self.y + other.y
        return Vector(xx, yy)
    def __repr__(self):
        return f"-->[{self.x}, {self.y}]"
```

```python
class ZeroModel:
    def __init__(self):
        self.m = 0
        self.c = 0

    def fit(self, data):
        # Calculate the means of x and y
        x_mean = sum([point[0] for point in data]) / len(data)
        y_mean = sum([point[1] for point in data]) / len(data)

        # Calculate the numerator and denominator for the slope (m)
        numerator = sum([(point[0] - x_mean) * (point[1] - y_mean) for point in data])
        denominator = sum([(point[0] - x_mean) ** 2 for point in data])

        # Calculate slope (m) and intercept (c)
        self.m = numerator / denominator
        self.c = y_mean - self.m * x_mean

    def predict(self, x):
        return self.m * x + self.c

# Example usage:
z = ZeroModel()
z.fit([(1, 2), (3, 4)])
print(z.predict(10))  # Output will be m*10 + c
```

```python
class ZeroModel:
    def __init__(self):
        self.m = 0
        self.c = 0

    def fit(self, point1, point2):
        """
        Learns the slope (m) and intercept (c) from two points.

        Args:
            point1: A tuple (x1, y1) representing the first point.
            point2: A tuple (x2, y2) representing the second point.
        """
        # Extract coordinates
        x1, y1 = point1
        x2, y2 = point2

        # Calculate slope (m)
        self.m = (y2 - y1) / (x2 - x1)

        # Calculate intercept (c)
        self.c = y1 - self.m * x1

    def predict(self, x):
        """
        Predicts the y value for a given x using the learned slope and intercept.

        Args:
            x: The x value for which to predict the y value.

        Returns:
            The predicted y value.
        """
        return self.m * x + self.c

    def __repr__(self):
        return f"(m,c) =>({self.m},{self.c})"

# Example usage:
z = ZeroModel()
point1 = (1, 5)
point2 = (2, 9)
z.fit(point1, point2)
#print(f"Slope: {z.m}, Intercept: {z.c}")

x_new = 5
predicted_y = z.predict(x_new)
print(f"Predicted y for x = {x_new}: {predicted_y}")
```
