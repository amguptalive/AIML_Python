```python
import matplotlib.image as image
import matplotlib.pyplot as plt

# Path to the image file
path = 'istockphoto-2155904673-1024x1024.jpg'

# Read the image
myimg = image.imread(path)

# Get the dimensions of the image
height, width, _ = myimg.shape

# Create an empty 2D array for the grayscale image using a simple for loop
gray_image = []
for i in range(height):
    gray_image.append([0] * width)

# Convert to grayscale by averaging the RGB values
for i in range(height):
    for j in range(width):
        r = myimg[i, j][0]  # Get the red value
        g = myimg[i, j][1]  # Get the green value
        b = myimg[i, j][2]  # Get the blue value
        gray_image[i][j] = (r + g + b) / 3  # Average the RGB values

# Display the grayscale image
plt.imshow(gray_image, cmap='gray')
plt.show()

# Print the shape of the grayscale image
print(len(gray_image), len(gray_image[0]))
```

```python
def distance(point1, point2):
    x1, y1 = point1
    x2, y2 = point2
    
    dx, dy = x1 - x2, y1 - y2
    
    return math.sqrt(dx*dx + dy*dy)
```

```python
point1 = (0, 1)
point2 = (10, 5)
distance(point1, point2)
```

```python
point1 = [0, 1]
point2 = [10, 5]
distance(point1, point2)
```

```python
def find_intersection(rect1, rect2):
    # Extract coordinates
    (x1_min, y1_min), (x1_max, y1_max) = rect1
    (x2_min, y2_min), (x2_max, y2_max) = rect2
    
    # Calculate the coordinates of the intersection rectangle
    x_inter_min = max(x1_min, x2_min)
    y_inter_min = max(y1_min, y2_min)
    x_inter_max = min(x1_max, x2_max)
    y_inter_max = min(y1_max, y2_max)
    
    # Check if there is an intersection
    if x_inter_min < x_inter_max and y_inter_min < y_inter_max:
        # Return the coordinates in the correct order
        return ((x_inter_min, y_inter_min), (x_inter_max, y_inter_max))
    else:
        return None

# Example usage
rectangle1 = ((1, 2), (5, 4))
rectangle2 = ((2, 1), (6, 3))
intersection = find_intersection(rectangle1, rectangle2)
print("Intersection:", intersection)

For rectangle1 = ((1, 2), (5, 4)) and rectangle2 = ((2, 1), (6, 3)), the intersection is ((2, 2), (5, 3)).

This should now correctly represent the bottom-left and top-right corners of the intersection rectangle.
```

Q6: 

You have millions of strings in an list, you have to group the anagrams together.

Input: ["car", "art", "cat", "rat", "arc", "tar", "act"]

Expected Output:
arc, car
act, cat
art, rat, tar

```python
# Try the other approach
def group_anagrams(words):
    anagrams = {}
    for word in words:
        # Sort the word to create a key
        sorted_word = ''.join(sorted(word))
        # Add the word to the corresponding list in the dictionary
        if sorted_word in anagrams:
            anagrams[sorted_word].append(word)
        else:
            anagrams[sorted_word] = [word]
    
    # Extract the grouped anagrams
    grouped_anagrams = list(anagrams.values())
    return grouped_anagrams

# Example usage
words = ["car", "art", "cat", "rat", "arc", "tar", "act"]
result = group_anagrams(words)
for group in result:
    print(", ".join(group))
```
